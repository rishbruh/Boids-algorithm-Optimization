```python
df= pd.read_csv("boids vs fps.csv")
#df= pd.read_csv("area vs fps.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(df['area'], df['fps dbscan'], color='b', label='DBSCAN')

plt.plot(df['area'], df['fps'], color='r', label='Original')

plt.plot(df['area'], df['HDBSCAN'], color='y', label='HDBSCAN')

plt.xlabel('Area of Environment (pixel^2)')

plt.ylabel('FPS')
 
plt.title('FPS v/s Area of environment')

plt.legend()
plt.show()

```

```python
#for boids v/s fps
import matplotlib.pyplot as plt

plt.plot(df['boids'], df['fps-DBSCAN'], color='b', label='DBSCAN')

plt.plot(df['boids'], df['fps'], color='r', label='Original')

plt.plot(df['boids'], df['fps-HDBSCAN'], color='y', label='HDBSCAN')

plt.xlabel('Number of Boids')

plt.ylabel('FPS')
 
plt.title('FPS v/s Number of Boids')

plt.legend()
plt.show()
```
